###  Theoretical Concepts

1. **Random Variable:** A function mapping outcomes of a random process to numbers.  
2. **Types:** Discrete (countable outcomes), Continuous (infinite possible values).  
3. **Discrete vs Continuous:** Discrete is countable (e.g., dice), Continuous can take any value (e.g., height).  
4. **PDF:** Describes likelihood of a random variable taking certain values.  
5. **CDF vs PDF:** CDF gives cumulative probability; PDF gives point probability.  
6. **Discrete Uniform:** Equal probability for all discrete outcomes.  
7. **Bernoulli Distribution:** Two outcomes (0 or 1) with p(success)=p, p(failure)=1-p.  
8. **Binomial Distribution:** Probability of k successes in n trials.  
9. **Poisson:** Events per fixed interval; used in traffic, call centers.  
10. **Continuous Uniform:** Equal likelihood across a range.  
11. **Normal Distribution:** Bell curve, symmetric, defined by mean & std dev.  
12. **Standard Normal:** Mean=0, Std=1; used for standardization.  
13. **Central Limit Theorem (CLT):** Sample means tend to normality with large samples.  
14. **CLT & Normal:** Enables normal approximation for inference.  
15. **Z-Statistics:** Standardized score for hypothesis testing.  
16. **Z-Score:** (X - mean) / std_dev; measures deviation from mean.  
17. **Point vs Interval Estimates:** Single value vs range of likely values.  
18. **Confidence Interval:** Range that likely includes population parameter.  
19. **Z-score & CI:** Z-scores determine CI boundaries.  
20. **Z for Comparing:** Standardizes different distributions.  
21. **CLT Assumptions:** Independence, large sample, finite mean/variance.  
22. **Expected Value:** Long-term average of a random variable.  
23. **Expected Value & Distribution:** Mean weighted by probabilities.  


```python
# Practical Implementation ===

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, poisson, uniform, norm, zscore
from scipy.stats import kurtosis, skew
import seaborn as sns

# 1. Generate a random variable
x = np.random.randint(1, 100)
print("Random variable:", x)

# 2. Discrete uniform distribution PMF
values = np.arange(1, 7)
pmf = np.ones_like(values) / len(values)
plt.bar(values, pmf)
plt.title("Discrete Uniform PMF")
plt.show()

# 3. Bernoulli PDF
def bernoulli_pdf(p):
    return {0: 1 - p, 1: p}
print("Bernoulli PDF (p=0.5):", bernoulli_pdf(0.5))

# 4. Binomial histogram
bin_data = binom.rvs(n=10, p=0.5, size=1000)
plt.hist(bin_data, bins=11, density=True)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()

# 5. Poisson distribution
pois_data = poisson.rvs(mu=3, size=1000)
sns.histplot(pois_data, kde=False)
plt.title("Poisson Distribution (mu=3)")
plt.show()

# 6. CDF of discrete uniform
cdf = np.cumsum(pmf)
plt.step(values, cdf)
plt.title("CDF of Discrete Uniform")
plt.show()

# 7. Continuous uniform
data_uniform = uniform.rvs(loc=0, scale=10, size=1000)
sns.histplot(data_uniform, kde=True)
plt.title("Continuous Uniform Distribution")
plt.show()

# 8. Normal distribution simulation
normal_data = np.random.normal(loc=50, scale=10, size=1000)
plt.hist(normal_data, bins=30, density=True)
plt.title("Normal Distribution Histogram")
plt.show()

# 9. Z-score and plot
zscores = zscore(normal_data)
sns.histplot(zscores, kde=True)
plt.title("Z-Scores of Normal Data")
plt.show()

# 10. CLT Simulation
samples = [np.mean(np.random.exponential(scale=2, size=30)) for _ in range(1000)]
sns.histplot(samples, kde=True)
plt.title("CLT from Exponential Data (Sample Size=30)")
plt.show()
```

```python
# Additional Practical Tasks ===

# 1. Multiple CLT samples
samples = [np.mean(np.random.normal(loc=0, scale=1, size=50)) for _ in range(1000)]
sns.histplot(samples, kde=True)
plt.title("CLT Verification with Normal Samples")
plt.show()

# 2. Standard normal plot
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)
plt.plot(x, y)
plt.title("Standard Normal Distribution")
plt.show()

# 3. Binomial probability calculation
p = binom.pmf(k=3, n=10, p=0.5)
print("P(X=3) for Binomial(n=10, p=0.5):", p)

# 4. Z-score for data point
x = 80
mean = 70
std_dev = 10
z = (x - mean) / std_dev
print("Z-score:", z)

# 5. Hypothesis test using Z
z = (80 - 75) / (10 / np.sqrt(30))
print("Z-statistic:", z)

# 6. Confidence interval
data = np.random.normal(100, 15, 100)
mean = np.mean(data)
sem = np.std(data, ddof=1) / np.sqrt(len(data))
ci = norm.interval(0.95, loc=mean, scale=sem)
print("Confidence Interval:", ci)

# 7. Confidence interval interpretation
print(f"The 95% confidence interval for the mean is {ci[0]:.2f} to {ci[1]:.2f}")

# 8. PDF of normal distribution
x = np.linspace(60, 140, 1000)
pdf = norm.pdf(x, 100, 15)
plt.plot(x, pdf)
plt.title("Normal Distribution PDF")
plt.show()

# 9. CDF of Poisson
x = np.arange(0, 20)
cdf = poisson.cdf(x, mu=3)
plt.step(x, cdf)
plt.title("CDF of Poisson (mu=3)")
plt.show()

# 10. Continuous uniform expected value
data = uniform.rvs(loc=5, scale=10, size=1000)
expected_val = np.mean(data)
print("Expected Value (Uniform):", expected_val)

# 11. Compare standard deviations
data1 = np.random.normal(100, 10, 1000)
data2 = np.random.normal(100, 20, 1000)
plt.hist(data1, alpha=0.5, label="SD=10")
plt.hist(data2, alpha=0.5, label="SD=20")
plt.legend()
plt.title("Comparison of Standard Deviations")
plt.show()

# 12. Range and IQR
data = np.random.normal(50, 10, 1000)
iqr = np.percentile(data, 75) - np.percentile(data, 25)
data_range = np.ptp(data)
print("Range:", data_range, "IQR:", iqr)

# 13. Z-score normalization
normalized = zscore(data)
sns.histplot(normalized, kde=True)
plt.title("Z-score Normalization")
plt.show()

# 14. Skewness and Kurtosis
print("Skewness:", skew(data))
print("Kurtosis:", kurtosis(data))
```
